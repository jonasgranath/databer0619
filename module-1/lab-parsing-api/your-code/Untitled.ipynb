```python
url_btc = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart'
param_dict = { 'id':'bitcoin', 'vs_currency':'eur', 'days':10}

```

```python
result_btc = requests.get(url_btc, params = param_dict) 

j_btc = result_btc.json()

df = pd.DataFrame(j_btc['prices'], columns=['time', 'price'])
df['time'] = pd.to_datetime(df['time'], unit = 'ms' )

df.set_index('time', inplace=True)

df.head(10)
```

```python
df.plot()
```
